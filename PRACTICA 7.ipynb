```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataset Iris
df = sns.load_dataset('iris')

# Gráfico de barras: promedio de largo y ancho de pétalos por especie
df.groupby('species')[['petal_length', 'petal_width']].mean().plot(kind='bar', figsize=(8,5))
plt.title("Promedio de Largo y Ancho de los Pétalos por Especie")
plt.xlabel("Especie")
plt.ylabel("Medida en cm")
plt.show()

# Histograma de la distribución de los largos de los sépalos
plt.hist(df['sepal_length'], bins=10, edgecolor='black')
plt.title("Distribución de Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frecuencia")
plt.show()

# Gráfico de dispersión: relación largo y ancho del pétalo
sns.scatterplot(x=df['petal_length'], y=df['petal_width'], hue=df['species'])
plt.title("Relación entre Largo y Ancho del Pétalo")
plt.show()

# Boxplot para comparar distribución de medidas del sépalo por especie
sns.boxplot(x=df['species'], y=df['sepal_length'])
plt.title("Distribución de Sepal Length por Especie")
plt.show()

# Gráfico de líneas: Variación de petal_length en el dataset
plt.plot(df.index, df['petal_length'])
plt.title("Variación del Largo del Pétalo")
plt.xlabel("Índice")
plt.ylabel("Petal Length")
plt.show()
```
