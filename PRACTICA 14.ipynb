```python
import pandas as pd

# Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Cargar el archivo
ruta_archivo = '/content/drive/MyDrive/nls97.csv'
df = pd.read_csv(ruta_archivo)

# Mostrar nombres de las columnas
print("\n--- Verificando nombres de columnas ---")
print(df.columns)

# --- Ejercicio 01: Errores Tipográficos y Valores Inconsistentes ---

# Ver valores únicos antes de la corrección
print("\n--- Ejercicio 01: Errores Tipográficos y Valores Inconsistentes ---")
print("Valores únicos en maritalstatus antes de la corrección:")
print(df['maritalstatus'].unique())

# Corrección de valores inconsistentes en maritalstatus
df['maritalstatus'] = df['maritalstatus'].replace({
    'Never-married': 'Single'
})

# Mostrar la frecuencia de valores después de la corrección
print("\nFrecuencia de valores después de la corrección:")
print(df['maritalstatus'].value_counts())

# --- Ejercicio 02: Errores Tipográficos y Valores Inconsistentes ---

print("\n--- Ejercicio 02: Errores Tipográficos y Valores Inconsistentes ---")
print("Valores únicos en highestdegree antes de la corrección:")
print(df['highestdegree'].unique())

# Corrección de valores inconsistentes en highestdegree
df['highestdegree'] = df['highestdegree'].replace({
    '2. High School': 'high school',
    '0. None': 'none',
    '4. Bachelors': 'bachelors',
    '3. Associates': 'associates',
    '5. Masters': 'masters',
    '1. GED': 'ged',
    '7. Professional': 'professional',
    '6. PhD': 'phd',
    'Unknown': 'unknown'
})

# Mostrar los valores únicos después de la corrección
print("\nValores únicos en highestdegree después de la corrección:")
print(df['highestdegree'].unique())

# --- Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas ---

print("\n--- Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas ---")

# Agrupar estados civiles poco frecuentes en una categoría
df['marital_grouped'] = df['maritalstatus'].replace({
    'Widowed': 'Single/No Partner'
})

# Mostrar valores únicos y frecuencia
print("\nValores únicos en marital_grouped:")
print(df['marital_grouped'].unique())

print("\nFrecuencia de valores en marital_grouped:")
print(df['marital_grouped'].value_counts())

# --- Ejercicio 04: Tipos de Datos ---

print("\n--- Ejercicio 04: Tipos de Datos ---")

# Ver tipo de datos antes de la conversión
print("\nTipo de datos en gender antes de la conversión:")
print(df['gender'].dtype)
print("Bytes ocupados antes de la conversión:", df['gender'].memory_usage(deep=True))

# Convertir 'gender' a tipo category para optimizar memoria
df['gender'] = df['gender'].astype('category')

# Ver tipo de datos después de la conversión
print("\nTipo de datos en gender después de la conversión:")
print(df['gender'].dtype)
print("Bytes ocupados después de la conversión:", df['gender'].memory_usage(deep=True))

# Guardar el dataframe limpio en Google Drive
ruta_salida = '/content/drive/MyDrive/nls97_cleaned.csv'
df.to_csv(ruta_salida, index=False)
print(f"\nArchivo limpio guardado en: {ruta_salida}")
```
