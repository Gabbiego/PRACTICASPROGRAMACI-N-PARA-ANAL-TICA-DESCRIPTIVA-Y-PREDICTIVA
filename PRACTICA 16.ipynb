```python
# Práctica #16: Datos Duplicados
# Nombre: Gabriela Gómez Ibarra
# Código: MIA000624B_Presencial_250206
# Fecha: 7 de marzo de 2025

# Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Importar librerías necesarias
import pandas as pd

# Cargar el archivo dirtydata.csv desde Drive
data = pd.read_csv('/content/drive/MyDrive/dirtydata.csv')

# Mostrar primeras filas e info del dataset
print("\nPrimeras filas del conjunto de datos:")
print(data.head())
print("\nInformación general del conjunto de datos:")
print(data.info())

# Ejercicio 01. Datos Perdidos
print("\n--- Ejercicio 01. Datos Perdidos ---")
print("\nCantidad de datos perdidos por columna:")
print(data.isnull().sum())

# Imputar valores faltantes en Calories con la media
calories_mean_before = data['Calories'].mean()
print(f"\nMedia de Calories antes de imputar: {calories_mean_before}")

data['Calories'] = data['Calories'].fillna(calories_mean_before)

calories_mean_after = data['Calories'].mean()
print(f"Media de Calories después de imputar: {calories_mean_after}")

# Ejercicio 02. Duplicidad Parcial en Duration
print("\n--- Ejercicio 02. Duplicidad Parcial Una columna (Duration) ---")
print(f"\nCantidad de registros duplicados en Duration: {data['Duration'].duplicated().sum()}")

# Conteo de valores únicos en Duration
print("\nConteo de valores únicos en Duration:")
print(data['Duration'].value_counts())

# Eliminar duplicados en Duration
data_no_duplicates_duration = data.drop_duplicates(subset=['Duration'])

# Ejercicio 03. Duplicidad en Pulse y Maxpulse
print("\n--- Ejercicio 03. Duplicidad Parcial Dos columnas (Pulse y Maxpulse) ---")
print(f"\nCantidad de registros duplicados en Pulse y Maxpulse: {data.duplicated(subset=['Pulse', 'Maxpulse']).sum()}")

# Conteo de combinaciones únicas
print("\nConteo de combinaciones únicas en Pulse y Maxpulse:")
print(data.groupby(['Pulse', 'Maxpulse']).size().reset_index(name='count'))

# Eliminar duplicados y quedarse con la última ocurrencia
data_no_duplicates_pulse_maxpulse = data.drop_duplicates(subset=['Pulse', 'Maxpulse'], keep='last')

# Ejercicio 04. Duplicidad Total
print("\n--- Ejercicio 04. Duplicidad Total o Exacta ---")
print(f"\nCantidad de registros con duplicidad exacta: {data.duplicated().sum()}")

data_last_exact = data.drop_duplicates(keep='last')

# Ejercicio 05. Cálculo de Media con y sin Duplicados
print("\n--- Ejercicio 05. Cálculo de la Media con Duplicados y Sin Duplicados ---")

# Calcular media con duplicados
pulse_mean_original = data['Pulse'].mean()
maxpulse_mean_original = data['Maxpulse'].mean()
print(f"\nMedia de Pulse con duplicados: {pulse_mean_original}")
print(f"Media de Maxpulse con duplicados: {maxpulse_mean_original}")

# Eliminar duplicados y calcular media sin duplicados
data_no_duplicates_pulse_maxpulse = data.drop_duplicates(subset=['Pulse', 'Maxpulse'], keep='last')

print(f"\nNúmero de filas después de eliminar duplicados en Pulse y Maxpulse: {len(data_no_duplicates_pulse_maxpulse)}")

pulse_mean_no_duplicates = data_no_duplicates_pulse_maxpulse['Pulse'].mean()
maxpulse_mean_no_duplicates = data_no_duplicates_pulse_maxpulse['Maxpulse'].mean()
print(f"Media de Pulse sin duplicados: {pulse_mean_no_duplicates}")
print(f"Media de Maxpulse sin duplicados: {maxpulse_mean_no_duplicates}")
```
