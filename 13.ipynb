```python
# CONECTAR GOOGLE DRIVE
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

# Ruta del archivo en tu Google Drive (ajusta según tu ubicación)
ruta_archivo = '/content/drive/MyDrive/miaad-nyc-r-s.csv'

# PRACTICA 13: CARGA DE DATOS
print("\n--- Práctica 13: Carga de Datos ---")
try:
    df = pd.read_csv(ruta_archivo, engine="python")  # Se eliminó skipfooter=1 por compatibilidad
    print("Archivo cargado correctamente.")
    print("Primeras 3 filas:\n", df.head(3))
except Exception as e:
    print(f"Error al cargar el archivo: {e}")

print("Últimas 3 filas (para verificar datos):\n", df.tail(3))
print("Dimensiones del dataframe (filas, columnas):", df.shape)

# PRACTICA 14: AGREGAR UNA COLUMNA
print("\n--- Práctica 14: Agregar una Columna ---")
print("Nombres originales de las columnas:\n", df.columns.tolist())

if df.columns[0].lower() != "index_miiad":
    df.rename(columns={df.columns[0]: "index_miiad"}, inplace=True)

df["nueva_columna"] = "valor_por_defecto"

print("Nombres corregidos de las columnas:\n", df.columns.tolist())

# PRACTICA 15: NORMALIZACIÓN DE NOMBRES DE COLUMNAS
print("\n--- Práctica 15: Normalización de Nombres de Columnas ---")
df.columns = df.columns.str.lower().str.replace(" ", "_", regex=True)
print("Nombres de columnas normalizados:\n", df.columns.tolist())

# PRACTICA 16: ERRORES EN TIPOS DE DATOS
print("\n--- Práctica 16: Errores en Tipos de Datos ---")
print("Tipos de datos antes de la conversión:\n", df.dtypes)

columnas_a_convertir = ["sale_price", "land_square_feet", "gross_square_feet"]
for col in columnas_a_convertir:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

print("Tipos de datos después de la conversión:\n", df.dtypes)

# PRACTICA 17: DICCIONARIO DE DATOS
print("\n--- Práctica 17: Diccionario de Datos ---")
diccionario_datos = {col: str(df[col].dtype) for col in df.columns}
print("Diccionario de datos:\n", diccionario_datos)
```
