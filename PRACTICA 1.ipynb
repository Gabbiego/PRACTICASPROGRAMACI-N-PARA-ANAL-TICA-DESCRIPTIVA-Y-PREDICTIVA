```python
import os
import re

# Ruta del archivo en Google Colab o entorno local
def obtener_ruta_archivo():
    ruta = "/mnt/data/mbox.txt"  # Archivo ya subido
    if not os.path.exists(ruta):
        from google.colab import files
        print("El archivo no fue encontrado. Sube 'mbox.txt'.")
        uploaded = files.upload()
        if "mbox.txt" in uploaded:
            return "mbox.txt"
        else:
            raise FileNotFoundError("No se encontró el archivo después de la carga.")
    return ruta

filename = obtener_ruta_archivo()

# Lista para almacenar los valores extraídos
spam_confidences = []

# Leer el archivo y extraer datos
try:
    with open(filename, "r", encoding="utf-8") as file:
        for line in file:
            match = re.search(r"X-DSPAM-Confidence: ([0-9.]+)", line)
            if match:
                spam_confidences.append(float(match.group(1)))

    # Calcular métricas
    if spam_confidences:
        count = len(spam_confidences)
        total = sum(spam_confidences)
        average = total / count
        variance = sum((x - average) ** 2 for x in spam_confidences) / count

        print("Archivo procesado correctamente.")
        print("Cantidad de datos extraídos:", count)
        print("Sumatoria de los datos:", total)
        print("Promedio de los datos:", average)
        print("Varianza de los datos:", variance)
    else:
        print("No se encontraron datos de X-DSPAM-Confidence en el archivo.")

except Exception as e:
    print("Error:", str(e))
```
